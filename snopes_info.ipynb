```python
import pandas as pd
from dateutil import parser



# Read the cleaned results and the results with justification
justified_df = pd.read_csv("snopes_results_with_justification.csv")



# Function to identify unique date formats in the claim_date column
def find_unique_date_formats(dates):
    unique_formats = set()
    for date in dates:
        try:
            parsed_date = parser.parse(date)
            unique_formats.add(parsed_date.strftime('%Y-%m-%d'))  # Standard format
        except (ValueError, TypeError):
            unique_formats.add("Invalid date format")
    return unique_formats

# Extract claim_date column and find unique formats
claim_dates = justified_df['claim_date'].dropna().unique()
unique_date_formats = find_unique_date_formats(claim_dates)




print(f"Unique date formats found in claim_date:{len(claim_dates)}")
for fmt in claim_dates:
    print(fmt)
```
